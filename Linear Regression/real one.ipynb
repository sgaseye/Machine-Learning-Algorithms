```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn import preprocessing
```

```python
data_train = pd.read_csv("Train/Train_Data.csv")
```

```python
empty = np.where(pd.isnull(data_train))
print(empty[0].shape)
coloumns_name = data_train.columns
print(coloumns_name[0])
```

```python
for i in range(empty[0].shape[0]):
    for j in range(empty[1].shape[0]):
        row_no = empty[0][i]
        col_no = empty[1][j]
        col = coloumns_name[col_no]
        data_train.at[row_no,col] = -1
        
```

```python
empty2 = np.where(pd.isnull(data_train))
print(empty2)
```

```python
for col_name in data_train.columns:
    le= preprocessing.LabelEncoder()
    print(col_name)
    data_train[col_name] = le.fit_transform(data_train[col_name])
    print(1)
```

```python
data_train
```

```python
data_train.shape
data_train = data_train.values
```

```python
Xt = data_train[:,1:-1]
Yt = data_train[:,-1]
```

```python
print(Xt.shape)
print(Yt.shape)
```

```python
x_train = Xt[:700]
x_val = Xt[700:]
y_train = Yt[:700]
y_val = Yt[700:]
```

```python
print(x_train.shape,y_train.shape)
print(x_val.shape,y_val.shape)
```

```python
Dense?
```

```python
from keras import models
from keras.layers import Dense
```

```python
model = models.Sequential()
model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation = 'relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='Adam',loss = 'mean_squared_error',metrics=['mae'])
```

```python
hist = model.fit(x_train,y_train,epochs = 150,batch_size=100,validation_data=(x_val,y_val))
```

```python
model.evaluate(x_train,y_train)
```
