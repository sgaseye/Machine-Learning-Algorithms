```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
a=np.array([1,2,3])
b=np.array([4,5,6,7])
a,b= np.meshgrid(a,b)#this func. create 2 matrices ->> a is repeated (z=no. of elment in b) z times along the row 
                     #and similarly ->> b is repeated (g=no. of elment in a) g times but along the column 
print(a)
print(b)
```

```python
fig=plt.figure()
axes =fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap='coolwarm')#red define high value region and blue define low value region
plt.show()
```

```python
fig=plt.figure()
axes =fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')#red define high value region and blue define low value region
plt.show()
```
