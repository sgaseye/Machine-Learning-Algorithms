```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(10)
y=(x-5)**2
print(x,y)
```

```python
plt.style.use ("seaborn")
plt.plot(x,y)
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

```python
X=0
lr=0.1
error=[]
plt.plot(x,y)
for i in range(50):#lets say we go 50 steps downward
    grad=2*(X-5)
    X= X- lr*grad
    Y=(X-5)**2
    error.append(Y)
    plt.scatter(X,Y)
    print(X)
```
