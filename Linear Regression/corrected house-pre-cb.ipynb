```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
features = ['feature_1', 'feature_2', 'feature_3', 'feature_4', 'feature_5','feature_6', 'feature_7', 'feature_8', 'feature_9', 'feature_10','feature_11', 'feature_12', 'feature_13']
X = boston.data
Y = boston.target


print(X.shape)
print(Y.shape)
```

```python
n = X.shape[1]
X = (X - X.mean())/X.std()
for i in range(n):
    plt.scatter(X[:, i],Y, color = 'blue',label = "X%d"%(i+1))
    plt.legend()
    plt.show()
```

```python
ones=np.ones((X.shape[0],1))
X=np.hstack((ones,X))
print(X.shape)
```

```python
print(X)
print(Y)
```

```python
def hypothesis(x, theta):
    return  theta[0]*x[0] + theta[1]*x[1] + theta[2]*x[2] + theta[3]*x[3] + theta[4]*x[4] +theta[5]*x[5] + theta[6]*x[6] + theta[7]*x[7] + theta[8]*x[8] + theta[9]*x[9] + theta[10]*x[10] + theta[11]*x[11] + theta[12]*x[12] + theta[13]*x[13]
def error(X,Y,theta):    
    m = X.shape[0]
    er = 0
    for i in range(m):
        hx = hypothesis(X[i], theta)
        er += (hx - Y[i])**2
        
    return er/m

def gradient(X,Y,theta):
    grad = np.zeros((14,))
    m,n = X.shape
    for i in range(m):
        hx = hypothesis(X[i],theta)
        grad[0] += (hx - Y[i])
        for j in range(n):
            grad[j] += (hx - Y[i])*X[i,j]
    return grad/m
    
def gradientDescent(X,Y,learning_rate=0.01):
    
    theta = np.zeros((14,))
    max_iter = 500
    error_list = []
    itr = 0
    
    while(itr<=max_iter):
         
        e = error(X,Y,theta)
        error_list.append(e)
        grad = gradient(X,Y,theta)
        theta = theta - learning_rate*grad
        itr+=1
        
    return theta, error_list
    
```

```python
final_theta, error_list = gradientDescent(X,Y)
```

```python
plt.plot(error_list)
plt.show()
```

```python
print(final_theta)
```

```python
from sklearn.metrics import r2_score
m = X.shape[0]
Y_pred = np.zeros(m)
for i in range(m):
    Y_pred[i] = hypothesis(X[i], final_theta)
accuracy = r2_score(Y, Y_pred)  
print("Training accuracy:", accuracy)
```
